```python
import matplotlib.pyplot as plt
import pandas as pd
import csv
import numpy as np
 

df = pd.read_csv('svd_psvd_truncate.csv', delimiter=';',decimal=',')
df = df[df.instance == "002_sg,txt"]

dps = df[df.method == "psvd_sparse"]
dps8 = dps[dps.cluster_states == 256]
dps12 = dps[dps.cluster_states == 4096]

ds = df[df.method == "svd"]
ds8 = ds[ds.cluster_states == 256]
ds12 = ds[ds.cluster_states == 4096]

psvd_8 = []
psvd_12 = []
svd_8 = []
svd_12 = []

for beta in [x / 10 for x in range(2, 11)]:
    dpsvd8 = dps8[dps8.β == beta]
    dsvd8 = ds8[ds8.β == beta]
    col_list_ps8 = dpsvd8['lnZ'].to_numpy()
    col_list_s8 = dsvd8['lnZ'].to_numpy()

    dpsvd12 = dps12[dps12.β == beta]
    dsvd12 = ds12[ds12.β == beta]
    col_list_ps12 = dpsvd12['lnZ'].to_numpy()
    col_list_s12 = dsvd12['lnZ'].to_numpy()
    psvd_8.append(np.std(col_list_ps8))
    psvd_12.append(np.std(col_list_ps12))
    svd_8.append(np.std(col_list_s8))
    svd_12.append(np.std(col_list_s12))
```

```python
xx = [x / 10 for x in range(2, 11)]
plt.scatter(xx, psvd_8, label = "psvd, 2^8")
plt.scatter(xx, psvd_12, label = "psvd, 2^12")
plt.scatter(xx, svd_8, label = "svd, 2^8")
plt.scatter(xx, svd_12, label = "svd, 2^12")
plt.legend()
plt.xlabel('beta')
plt.ylabel('std(ln(Z))')
# plt.xlim([0.1, 1.1])
plt.yscale("log")
plt.title("P4, instance 002")
plt.show()
```
